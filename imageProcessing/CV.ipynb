# Computer Vision Tutorial

The human eye and brain are readily able to identify the gist of images. However, they are not easily able to precisely determine what is happening in the image. For that, it's best to implement techniques that fall under the umbrella of "computer vision".  
  
It's a big field, and a common place to start is with lena.

```python
#first i need to read in the image
lena_img=imread('lena.jpg') 

#then i want to show the image
imshow(lena_img,origin='lower')
```

What can we say about this image?

-  We can say it is a bit over 512 pixels wide and 512 pixels tall.  
-  We can assume each pixel is a combination of Blue, Green, and Red, as that is the case for virtually all digital images.

While that may not seem like much, it alludes to something very fundamental: An image is a matrix (or "array"). The dimensions of the array define the width, height, and number of channels.
    
In the case of lena, that means:

```python
#get the shape of the image array
lena_img.shape 
```

As you might expect, lena is 512 pixels wide, 512 pixels tall, and 3 channels (Blue, Green, and Red) deep.

To look a little closer at the channels Let's pick one of the pixels and see what its Blue, Green, and Red values are:

```python
#print all the channels for the pixel at the intersection of the 200th row (the y-value) and 500th column (the x-value)
lena_img[200,500,:]
```

```python
This is telling us 4 things:
    
1. Each value is an 8-bit integer, meaning it has a value between 0 and 255 (see below)
2. The blue value is 238
3. The green value is 203
4. The red value is 165


```
