### An aside: the 8-bit integer

Everything in a computer is made up of $1s$ and $0s$. A single character (ie a $1$ or a $0$) is referred to as a *bit*.

Because a single bit can be either a $1$ or a $0$, it has $2^1=2$ possible combinations.

If you put two bits together there are $2^2=4$ combinatons. Three bits yield $2^3=8$ combinations and so on.

If you follow this through, you will see that an 8-bit value has $2^8=256$ combinations. This is why the Red, Green, and Blue values can range from 0 to 255... because they are each 8-bit integers!

### Color spaces: more than just RGB

*Caution: You could spend months learning about color theory, but you don't need to know everything in order to get started. If you want to dive deep you can check out the wikipedia page for any of the terms below*

While RGB is perhaps the most common color spaces, there are some others worth mentioning. 

- **CMYK**: Stands for Cyan, Magenta, Yellow, & Key(Black). This color space is more commonly used for mixing inks. It has almost no role in computer vision
- **BGR**: Stands for Blue, Green, & Red. This is the same model as RGB, the channels are just rearranged. For some reason, some cameras output data in BGR format
- **HSV**: Stands for Hue, Saturation, & Value. This is used frequently in computer vision
- **Lab**: Stands for Lightness, a, & b. This is also used freqently in computer vision.

### Analyzing a splatter

While lena is a standard example, she doesn't have too much low-hanging fruit from a product design perspective.

Instead, let's try to quantify how much blue paint is in the image below.

```python
splatter_img=imread('splatter.jpg')
imshow(splatter_img)
```

Unfortunately, before we can analyze the image, we need to convert it from an RGB colorspace (the default for iPython) to a BGR colorspace (the default for the `cv2` library)

```python
r,g,b=cv2.split(splatter_img)
splatter_img = cv2.merge([b,g,r])
```

Now we can do some manipulation

```python
#first let's create an HSV and Lab representation of the image
splatter_hsv=cv2.cvtColor(splatter_img,cv2.COLOR_BGR2HSV)
splatter_lab=cv2.cvtColor(splatter_img,cv2.COLOR_BGR2LAB)

#next, extract each channel from the BGR, HSV, and LAB color spaces

splatter_blue=splatter_img[:,:,0] #get all the rows, all the columns, and the first channel
splatter_green=splatter_img[:,:,1] #get all the rows, all the columns, and the second channel
splatter_red=splatter_img[:,:,2] #get all the rows, all the columns, and the third channel

splatter_h=splatter_hsv[:,:,0] #get all the rows, all the columns, and the first channel
splatter_s=splatter_hsv[:,:,1] #get all the rows, all the columns, and the second channel
splatter_v=splatter_hsv[:,:,2] #get all the rows, all the columns, and the third channel

splatter_l=splatter_lab[:,:,0] #get all the rows, all the columns, and the first channel
splatter_a=splatter_lab[:,:,1] #get all the rows, all the columns, and the second channel
splatter_b=splatter_lab[:,:,2] #get all the rows, all the columns, and the third channel 
#

#plotting stuff
sx=subplot(3,3,1)
sx.set_title("Channel: blue")
sx.imshow(splatter_blue)
sx=subplot(3,3,2)
sx.set_title("Channel: green")
sx.imshow(splatter_green)
sx=subplot(3,3,3)
sx.set_title("Channel: red")
sx.imshow(splatter_red)

sx=subplot(3,3,4)
sx.set_title("Channel:Hue")
sx.imshow(splatter_h)
sx=subplot(3,3,5)
sx.set_title("Channel:Saturation")
sx.imshow(splatter_s)
sx=subplot(3,3,6)
sx.set_title("Channel:Value")
sx.imshow(splatter_v)

sx=subplot(3,3,7)
sx.set_title("Channel:Lightness")
sx.imshow(splatter_l)
sx=subplot(3,3,8)
sx.set_title("Channel:A")
sx.imshow(splatter_a)
sx=subplot(3,3,9)
sx.set_title("Channel:B")
sx.imshow(splatter_b)
```

There are a few things worth pointing out above.

- The images are all shown in greyscale. This is because they are all single channel arrays, meaning each pixel has only one value (between 0 and 255). The images are scaled where black is the lowest value in the image, and white is the highest value.
- The Red, Green, and Blue channels are pretty useless. RGB (and BGR) is convenient for *producing* and *capturing* light, but it doesn't mimic the way humans see very well, so analysis of the images is difficult in this color space.
- The Value and Lightness channels are similarly unhelpful. These are essentially a measure of brightness, and not much information comes out of that.
- The Hue channel is mangled. Not sure why, but this seems to happen often. I don't use it.
- The Saturation, A, and B channels show promise.

At this point, we could probably either Saturation, A, or B (these are often the most valuable channels). I am going to choose Saturation.

There are a few things we can do to analyze this image, but the first thing is to threshold it so that our image becomes an array of $1s$ and $0s$

```python
ret,thresh = cv2.threshold(splatter_s,0,1,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
imshow(thresh)
```

At this point we could just count up all the $1s$ in the array to see how much blue paint is in the image.

```python
#count all the white pixels
white=thresh.sum()

#calculate the percentage of white pixels
percentage=float(white)/thresh.size
print percentage
```

Great, so I can say that there is 4% covered by the image, but what if I wanted to count how many individual spots?

First, it would probably help to despeckle the image

```python
#calculate a despeckled image
despeckled=cv2.medianBlur(thresh,9)
imshow(despeckled)
```

```python
labeled, nr_blobs=ndimage.label(despeckled)
imshow(labeled)
```

```python
centers=[]
for i in xrange(nr_blobs):
    temp=labeled==i+1
    temp2=temp.astype(np.uint8)
    moments=cv2.moments(temp2)
    centerY=moments['m01']/moments['m00']
    centerX=moments['m10']/moments['m00']
    centers.append((int(centerX),int(centerY)))
```

```python
for center in centers:
    cv2.circle(splatter_img,center,2,(255,0,0),thickness=-1)

imshow(splatter_img)
```
