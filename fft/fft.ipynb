# Using Fast Fourier Transforms to calculate frequency of sampled data

Often when doing experiments in the lab, it is valuable to be able to calculate the parameters of noisy data. This can be done using Fast Fourier Transforms.  
In order to calculate it, you will need to have/know a few things:
    
- You will need to know your sampling frequency
- You will need to have access to a tool such as python (free and awesome) or matlab (incumbent)

That is all you need. However, before I can show you to calculate the frequency of the data, I'm going to create some noisy artificial data to analyze. This way you can see the that the sin waves used to generate the data matches what we get from the Fourier Transform.

Be sure to take note of the frequency, amplitude, and sampling rate values, as they define our signal.

```python
#creating some noisy data

t=linspace(0,1000,1000) #create some alues that represent time

frequency1=0.8
frequency2=0.6

amplitude1=4
amplitude2=2

samplingRate=100

noise=0.5

rand1=np.random.random_sample(size=(t.shape))# create an array of random number from 0 to 1
rand2=np.random.random_sample(size=(t.shape))# create another array of random numbers from 0 to 1
rand=(rand1-rand2)*noise #subtract the arrays to give me a new array of random numbers from -1 to 1, and add some noise

y=amplitude1*np.cos(2*pi*frequency1*t/samplingRate)-rand*10 + amplitude2*np.cos(2*pi*frequency2*t/samplingRate)-rand*10

plot(t,y,label="Noisy Data")
legend()
```

A few things to note about our signal:
    
- There are $1000$ samples
- There is some randomness to it that makes it noisy
- There are two sinusoids present:
    - The amplitdes are $4$ and $2$, respectively
    - The frequencies are $0.8$ and $0.2$ respectively.

So would we calculate these parameters directly from the raw data if we didn't alreadty know it? First we calculate the Fast Fourier Transform ([documentation](http://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.fft.html#numpy.fft.fft))

```python
#calculate the fast fourier transform
Y=fft.fft(y)
```

The transform takes the array of sample data (which existed in the time domain) and converts it to *bins* (which exist in the frequency domain). 

If we take a look at the transform we can see that it is comprised of complex numbers. This is in contrast to the sample data, which consisted of real numbers

```python
print Y
```

The next step is to calculate the magnitude and angle of each complex number in the Fourier Transform:

```python
Y_mag=np.abs(Y) #calculate the magnitude
Y_angle=np.angle(Y) #calculate the angle


#this is all just plotting stuff
sx=subplot(1,2,1)
plot(Y_mag, label='Magnitude of Fourier Transform')
leg=plt.legend()
sx=subplot(1,2,2)
plot(Y_angle, label='Angle of Fourier Transform')
leg=plt.legend()
```

Not terribly interesting at first glance, but it actually tells us a lot. There are a few things to notice here:
    
- The number of values (*bins* to be specific...we are still in the frequency domain) is $1000$, same as the sample data and Fourier Transform.
- The magnitude plot is symmetrical. This will always be true for most cases (it has been true for all my cases). This means we only need to look at half (I choose the left half) of the plot.
- There seems to be a lot happening at the edges, let's zoom into the left corner (ie, let's look a tthe first $100$ bins to get a better look at what's going on there.

```python
plot(Y_mag[0:100])
```

Interesting, There are two big spikes at the left edge.

The X positions of these spikes represent the frequencies (again, in *bins*) present in our signal

Great! So let's find the those frequencies.

```python
#this code implementation isn't the most efficient, but it should be pretty clear

Y_mag_half=Y_mag[:Y_mag.shape[0]/2] # we only need to look at the left half of the magnitude array

max1BinFrequency=argmax(Y_mag_half)# find the location of the biggest frequency
AMPLITUDE1=Y_mag_half[max1BinFrequency] #record the biggest frequency

Y_mag_half[max1BinFrequency]=0 #replace the max value with zero
max2BinFrequency=argmax(Y_mag_half) #searching for the max again will now return the second largest number in the array
AMPLITUDE2=Y_mag_half[max2BinFrequency] #record the second biggest frequency

print 'Max Value = {}, Secondary Max Value = {}'.format(max1BinFrequency,max2BinFrequency)
```

So our frequencies are is $8$ and $6$ *bins*... what is that in $Hz$?.

It turns out to be a straightforward rule: The number of bins is proprotional to the sampling frequency. Put another way:

```python
bins=Y.shape[0] #calculate the number of bins
hzPerBin=samplingRate/float(bins) #calculate the ratio of Hz to bins
print '1 bin = {} Hz'.format(hzPerBin)
```

Almost there, now we just need to multiply our new ratio by our number of bins, and that will tell us the frequency present in our signal

```python
calcFrequency1=max1BinFrequency*hzPerBin
calcFrequency2=max2BinFrequency*hzPerBin

```

We can then calculate the angles by looking them up using our max bin frequencies:

```python
calcAngle1=Y_angle[max1BinFrequency]
calcAngle2=Y_angle[max2BinFrequency]
```

Lastly we can calculate the amplitudes by dividing the values by half the numbe rof bins

```python
calcAmplitude1=AMPLITUDE1/(bins/2.)
calcAmplitude2=AMPLITUDE2/(bins/2.)
```

Printing the values shows that they do indeed match our input

```python
print 'Frequency1 = {} Hz, Angle1 = {} Radians, Amplitude = {}'.format(calcFrequency1,calcAngle1,calcAmplitude1)
print 'Frequency2 = {} Hz, Angle2 = {} Radians, Amplitude = {}'.format(calcFrequency2,calcAngle2,calcAmplitude2)
```

Lastly, plotting these calculated values over our original data verifies that our calculation is looking pretty good

```python
calculated=calcAmplitude1*np.cos(2*pi*calcFrequency1*t/samplingRate) + calcAmplitude2*np.cos(2*pi*calcFrequency2*t/samplingRate)


plot(t,y, label = 'Raw Data')
plot(t,calculated,'r-', lw=5, label='Calculated Waveform')
plt.legend()
```
